```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X = df[["Total Fat","Carbohydrates","Protein", "Sodium"]]
y = df["Calories"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model_regression = LinearRegression()
```

```python
model_regression.fit(X_train, y_train)
```

```python
y_pred = model_regression.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
coef = pd.DataFrame(model_regression.coef_, X.columns, ["Coefs"])
```

```python
coef
```

```python
# calories = 9 * fat + 4 * carbs + 4 * protein
```

```python
results = pd.DataFrame({"Real": y_test, "Predicted values": y_pred})
results = results.sample(n = 20)
results
```
